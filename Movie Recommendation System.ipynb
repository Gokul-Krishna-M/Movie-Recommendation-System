```python
Movie_Recommendation = Movie_Recommendation.fillna("")#Filling the missing values with null
Movie_Recommendation.isnull().sum()
```

```python
Movie_Recommendation
```

```python
Features_selected = ['genres','tagline','title','director','keywords']#Selecting the required Features
print(Features_selected)
```

```python
Movie_Recommendation_features_selected = Movie_Recommendation[Features_selected]#Creating a new data frame from the exisitng data frame with the selected features
Movie_Recommendation_features_selected
```

```python
#Combining the selected features
combined_features = Movie_Recommendation_features_selected['genres'] + Movie_Recommendation_features_selected['director'] + Movie_Recommendation_features_selected['tagline'] + Movie_Recommendation_features_selected['keywords'] + Movie_Recommendation_features_selected['title']
```

```python
combined_features.head()
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)#Vectorizing the combined features
display (feature_vectors.shape)
print (feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
print  (similarity )#Using the cosine similarity function to find the similarity between each movie
```

```python
display(similarity.shape)
```

```python
movie_name = input(' Enter your favourite movie name : ')
```

```python

list_of_all_titles = Movie_Recommendation ['title'].tolist()
print(list_of_all_titles)#Converting the title of movies to a list
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)#Using the difflib library to find the closest match of the title which user have entered
print(find_close_match)
```

```python
close_match = find_close_match[0]#Taking the first match as the input
print(close_match)
```

```python
index_of_the_movie = Movie_Recommendation[Movie_Recommendation.title == close_match]['index'].values[0]#Getting the index of the title
print(index_of_the_movie)
```

```python
similarity
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))#Getting the similarity of the movie of user input with every other movie and converting it into a list
print(similarity_score)
similarity_score
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) #Sorting the list in the descending order to get the most similar movies first
print(sorted_similar_movies)
```

```python
sorted_similar_movies = sorted_similar_movies[1:]
```

```python
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = Movie_Recommendation[Movie_Recommendation.index==index]['title'].values[0]
  if (i<31):
    print(i, '.',title_from_index)
    i+=1
```
